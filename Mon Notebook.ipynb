### Charger + pré-traiter les images, afin qu'elles respectent les specifications des images d'entrée en VGG16

```python
def reduce_dim_image(path_image):
    print(path_image)
    # Charger l'image et la redimensionner à la taille requise de 224×224 pixels.
    img = load_img(path_image, target_size=(224, 224))
    # Convertir les pixels en un tableau NumPy afin de pouvoir travailler avec dans Keras
    img = img_to_array(img)
    # Redimensionnement + couleur RBG=3
    img = img.reshape((3,224,224))
    img = np.expand_dims(img, 0)
    
    # Préparer de nouvelles entrées pour le réseau VGG16.
    img = preprocess_input(img)
    # obtenir une prédiction de la probabilité d'appartenance de l'image à chacun des 1000 types d'objets connus.
    return model.predict(img).flatten()
```
